```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
messages=re.split(pattern,data)
messages
```

```python
#now we are getting an empty string in first so to remove that we can do on thing 
messages=messages[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({'user_message':messages,'dates':dates})
```

```python
#now i want to convert the dAte format into datetime format 
df['date'] = pd.to_datetime(df['dates'], format='%d/%m/%y, %H:%M - ')
df
```

```python
users=[]
messages=[]
for message in df['user_message']:
    entry=message.split(':',1)
    if len(entry)==2:
        users.append(entry[0])
        messages.append(entry[1])
    else:
        users.append('group_notification')
        messages.append(entry[0])
    
```
